# Firstly load the 2 cleaned datasets here

```python
# Firstly load the 2 csv files 
import pandas as pd

# Load the 2 cleaned csv file 
cleaned_accident_csv_path = "../datasets/cleaned_acciddent.csv"
cleaned_node_csv_path = "../datasets/cleaned_node.csv"

cleaned_accident_df = pd.read_csv(cleaned_accident_csv_path)
cleaned_node_df = pd.read_csv(cleaned_node_csv_path)
```

```python
cleaned_accident_df.head(5)
```

```python
cleaned_node_df.head(5)
```

## Now we will merge the 2 datasets, merge key = "NODE_ID" to get more information related to node for each accident records 

```python
# Join these 2 dataframes with 2 keys: ACCIDENT_NO + NODE_ID, perform a left join as i want to keep all of the information 
# from the accident dataframe 
merged_df = pd.merge(cleaned_accident_df, cleaned_node_df, on=["ACCIDENT_NO", "NODE_ID"], how="left")
```

```python
# Display some first few rows of the dataframe
merged_df.head()
```

```python
# save this merged dataframe to a new csv file for future use --> no need to reload each time --> save resources
merged_df.to_csv('../datasets/merged.csv', index=False)
```
