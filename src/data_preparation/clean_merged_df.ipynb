# This file will analyze some really basic aspects of the merged.csv datasets

```python
# load the csv file 
import pandas as pd

merged_df = pd.read_csv("../../datasets/merged.csv")
```

```python
# display some first few rows again for overview
merged_df.head()
```

```python
# get the length of this dataset
print("The number of records of this merged dataset is: ", len(merged_df))
```

```python
# now i want to check the NaN value + value counts for this new datasets once again
all_features = list(merged_df.columns)

# function display NaN + value counts for all columns of a dataframe
def display_info(df, columns_to_display):
    for col in columns_to_display:
        print("This is column: " + col)
        print(df[col].value_counts(dropna=False))
        print("=" * 80)

display_info(merged_df, all_features)
```

```python
# As the number of NaN values is < 0.01% of the dataset total records so i will drop all of the records with NaN values 
cleaned_merged_df = merged_df.dropna()

# and then save / replace this cleaned_merged_df to the current merged.csv
cleaned_merged_df.to_csv("../../datasets/merged.csv", index=False)
```
